## Exemplo Resolvido 1: Tabuada

**O problema:** Crie um programa que peça um número ao usuário e imprima a tabuada desse número de 1 a 10 usando o `for`.

```python
numero = int(input("Você quer a tabuada de qual número? "))

print(f"\n--- Tabuada do {numero} ---")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i:2d} = {resultado:2d}")
```

*Dica: O formato `{:2d}` dentro da f-string diz para o Python reservar 2 espaços na hora de imprimir aquele número. Isso ajuda a deixar os números alinhados bonitinhos na tela!*

## Exemplo Resolvido 2: Somatória com While

**O problema:** Peça números ao usuário repetidamente. O programa deve somar todos os números digitados e só deve parar (e mostrar o total) quando o usuário digitar o **número ZERO**.

```python
soma = 0
while True:
    valor = float(input("Digite um valor para somar (ou 0 para parar): "))

    if valor == 0:
        break # Sai do loop quando o usuário digitar zero

    soma = soma + valor

print(f"\nA soma total foi de: {soma}")
```

### Desafio 3.
Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

```python
A=80000
B=200000
taxa_a = 0.03
taxa_b = 0.015
anos = 0

while A < B:
    A = A * (1 + taxa_a)
    B = B * (1 + taxa_b)
    anos += 1

print(f"Serão necessários {anos} anos para que a população do país A ultrapasse ou iguale a população do país B.")
```

### Desafio 4.
Altere o programa anterior permitindo ao usuário informar as populações e as taxas de crescimento iniciais. Valide a entrada e permita repetir a operação.

```python
def calcular_anos():
    while True:
        try:
            pop_a = int(input("Digite a população inicial do país A: "))
            taxa_a = float(input("Digite a taxa de crescimento anual do país A (ex: 0.03 para 3%): "))
            pop_b = int(input("Digite a população inicial do país B: "))
            taxa_b = float(input("Digite a taxa de crescimento anual do país B (ex: 0.015 para 1.5%): "))

            if pop_a <= 0 or pop_b <= 0 or taxa_a < 0 or taxa_b < 0:
                print("As populações e taxas de crescimento devem ser números positivos. Tente novamente.")
                continue
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite números válidos.")

    anos = 0
    if pop_a > pop_b:
        print("A população inicial de A já é maior ou igual à de B. Não são necessários anos para ultrapassar.")
        return

    while pop_a < pop_b:
        pop_a *= (1 + taxa_a)
        pop_b *= (1 + taxa_b)
        anos += 1

    print(f"Serão necessários {anos} anos para que a população do país A ultrapasse ou iguale a população do país B.")

while True:
    calcular_anos()
    continuar = input("Deseja calcular novamente? (s/n): ").lower()
    if continuar != 's':
        break
```

### Desafio 22.
Altere o programa de cálculo dos números primos, informando, caso o número não seja primo, por quais número ele é divisível.

```python
while True:
    try:
        numero = int(input("Digite um número inteiro: "))
        if numero < 0:
            print("Por favor, digite um número inteiro não-negativo.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

divisores = []

if numero <= 1:
    print(f"{numero} não é um número primo.")
    if numero == 1:
        print("Divisível por: 1")
    elif numero == 0:
        print("O zero não tem divisores próprios no contexto de números primos.")
else:
    for i in range(1, numero + 1):
        if numero % i == 0:
            divisores.append(i)

    if len(divisores) == 2:
        print(f"{numero} é um número primo.")
    else:
        print(f"{numero} não é um número primo.")
        print(f"Divisível por: {', '.join(map(str, divisores))}")
```

### Desafio 23.
Faça um programa que mostre todos os primos entre 1 e N sendo N um número inteiro fornecido pelo usuário. O programa deverá mostrar também o número de divisões que ele executou para encontrar os números primos. Serão avaliados o funcionamento, o estilo e o número de testes (divisões) executados.

```python
def is_prime(num, divisions_count):
    if num <= 1:
        return False, divisions_count
    if num <= 3:
        return True, divisions_count
    if num % 2 == 0 or num % 3 == 0:
        divisions_count += 2 # For checking divisibility by 2 and 3
        return False, divisions_count

    i = 5
    while i * i <= num:
        divisions_count += 2 # For checking divisibility by i and i + 2
        if num % i == 0 or num % (i + 2) == 0:
            return False, divisions_count
        i += 6
    return True, divisions_count


while True:
    try:
        N = int(input("Digite um número inteiro positivo N: "))
        if N > 0:
            break
        else:
            print("Por favor, digite um número inteiro positivo.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

print(f"\nNúmeros primos entre 1 e {N}:")
primos_encontrados = []
total_divisoes = 0

for numero in range(1, N + 1):
    e_primo, divisoes_atuais = is_prime(numero, 0)
    total_divisoes += divisoes_atuais
    if e_primo:
        primos_encontrados.append(numero)

print(primos_encontrados)
print(f"Total de divisões executadas: {total_divisoes}")
```

### Desafio 25.
Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da turma varia entre 0 e 25,26 e 60 e maior que 60; e então, dizer se a turma é jovem, adulta ou idosa, conforme a média calculada.

```python
while True:
    try:
        N = int(input("Digite o número de pessoas na turma (N): "))
        if N <= 0:
            print("Por favor, digite um número inteiro positivo para N.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

soma_idade = 0
for i in range(1, N + 1):
    while True:
        try:
            idade = int(input(f"Digite a idade da {i}ª pessoa: "))
            if idade < 0:
                print("A idade não pode ser negativa. Por favor, digite uma idade válida.")
                continue
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro para a idade.")
    soma_idade += idade

media_idade = soma_idade / N

print(f"\nA média de idade da turma é: {media_idade:.2f}")

if 0 <= media_idade <= 25:
    print("A turma é jovem.")
elif 26 <= media_idade <= 60:
    print("A turma é adulta.")
else:
    print("A turma é idosa.")
```

### Desafio 26.
Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
while True:
    try:
        total_eleitores = int(input("Digite o número total de eleitores: "))
        if total_eleitores <= 0:
            print("Por favor, digite um número inteiro positivo para o total de eleitores.")
            continue
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

votos_candidato1 = 0
votos_candidato2 = 0
votos_candidato3 = 0

print("\nCandidatos:\n1 - Candidato A\n2 - Candidato B\n3 - Candidato C")

for i in range(1, total_eleitores + 1):
    while True:
        try:
            voto = int(input(f"Eleitor {i}: Digite o número do seu candidato (1, 2 ou 3): "))
            if voto == 1:
                votos_candidato1 += 1
                break
            elif voto == 2:
                votos_candidato2 += 1
                break
            elif voto == 3:
                votos_candidato3 += 1
                break
            else:
                print("Voto inválido. Por favor, digite 1, 2 ou 3.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro (1, 2 ou 3).")

print("\n--- Resultado da Eleição ---")
print(f"Candidato A: {votos_candidato1} votos")
print(f"Candidato B: {votos_candidato2} votos")
print(f"Candidato C: {votos_candidato3} votos")
```

### Desafio 27.
Faça um programa que calcule o número médio de alunos por turma. Para isto, peça a quantidade de turmas e a quantidade de alunos para cada turma. As turmas não podem ter mais de 40 alunos.

```python
while True:
  try:
    quantidade_turmas = int(input("Digite a quantidade de turmas: "))
    if quantidade_turmas <= 0:
      print("Por favor, digite um numero inteiro positivo para a quantidade de turmas.")
      continue
    break # Break the inner while loop only if input is valid and positive
  except ValueError:
    print("Entrada invalida.")

total_alunos = 0
for i in range(1, quantidade_turmas + 1):
    while True:
        try:
            alunos_na_turma = int(input(f"Digite a quantidade de alunos para a turma {i} (máximo 40): "))
            if 0 < alunos_na_turma <= 40:
                total_alunos += alunos_na_turma
                break
            else:
                print("A quantidade de alunos deve ser entre 1 e 40.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

media_alunos_por_turma = total_alunos / quantidade_turmas
print(f"O número médio de alunos por turma é: {media_alunos_por_turma:.2f}")
```
